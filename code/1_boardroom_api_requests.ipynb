```python
dao_list = df['cname'].to_list()
```

```python
count = 0
proposals_lists = []
for i in dao_list:
    count += 1
    if count % 10 == 0:
        time.sleep(np.random.uniform(5, 10))
    req = requests.get(f'https://api.boardroom.info/v1/protocols/{i}/proposals')
    req_txt = req.text
    req_lst = ast.literal_eval(req_txt)
    proposals_lists.append(req_lst)
    
```

```python
proposal_dfs_list = []
for i in proposals_lists:
    df = pd.DataFrame(i['data'])
    proposal_dfs_list.append(df)
```

```python
proposal_dfs = pd.concat(proposal_dfs_list)
```

```python
proposal_dfs.to_excel('dao_proposals.xls')
```

```python
ref_id_list = proposal_dfs['refId'].to_list()
```

```python
len(ref_id_list)
```

```python
counter = 0
votes = []
for i in ref_id_list:
    counter += 1
    if counter % 10 == 0:
        time.sleep(np.random.uniform(3,5))
    vote = requests.get(f'https://api.boardroom.info/v1/proposals/{i}/votes')
    vote_txt = vote.text
    vote_lst = ast.literal_eval(vote_txt)
    votes.append(vote_lst)
```

```python
len(votes)
```

```python
votes_dfs_list = []
for i in votes:
    df = pd.DataFrame(i['data'])
    votes_dfs_list.append(df)
```

```python
vote_df = pd.concat(votes_dfs_list)
```

```python
vote_df.drop(columns= 'reason', inplace=True)
```
